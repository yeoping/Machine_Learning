## Import sympy, numpy, matplotlib

```python
import matplotlib.pyplot as plt 
import numpy as np
import sympy as sy
```

## Define a differentiable function : $f(x)=x^3+6x^2-1$


```python
def f(x):
    x=sy.symbols('x')
    f=x**3+6*x**2-1
    return f
```

## Define the first-order Taylor approximation at the selected point.  <u> (x is 2) 
<h3>: $\hat{f(x)}=36x-41$

```python
def taylor_f(x):
    taylor_f=sy.diff(f(x)).subs(x,2)*(x-2)+f(x).subs(x,2)
    return taylor_f
```

## Define a domain of the function , plot the function and mark the selected point (2,31)

```python
xx=np.linspace(-10,10,1000)
x=sy.symbols('x')
plt.plot(xx,sy.lambdify(x,f(x))(xx),'b',label='f(x)')
plt.scatter(2,31,marker='*')
plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0)
plt.show()
```

## Plot the Taylor approximation with the same domain of the original function.

```python
xx=np.linspace(-10,10,1000)
plt.plot(xx,sy.lambdify(x,f(x))(xx),'b',label='f(x)')
plt.scatter(2,31,marker='*')
plt.plot(xx,sy.lambdify(x,taylor_f(x))(xx),'r',label='taylor_f(x)')
plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0)
plt.show()
```
